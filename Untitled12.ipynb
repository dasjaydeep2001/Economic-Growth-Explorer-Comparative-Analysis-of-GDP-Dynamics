```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import plotly.express as px
import plotly.offline as pyo
import os
import plotly.io as pio
```

 **How many unique Countries Data we're having in the dataset?**

```python
unique_countries = df['Country Name'].nunique()

print("Number of unique countries:", unique_countries)
```

**How many years of Data we're having for India**

```python
years_for_india = df[df['Country Name'] == 'India']['Year'].nunique()

print("Number of years of GDP data for India:", years_for_india)
```

 **Whats the world GDP in 1969?**

```python
gdp_1969 = df[df['Year'] == 1969]['Value'].sum()

print("World GDP in 1969:", gdp_1969)
```

**Whats the % of GDP growth of India in Year 1994 (*as compare to 1993)?**

```python
gdp_1993 = df[(df['Country Name'] == 'India') & (df['Year'] == 1993)]['Value'].values[0]
gdp_1994 = df[(df['Country Name'] == 'India') & (df['Year'] == 1994)]['Value'].values[0]

# Calculate the percentage growth
percentage_growth = ((gdp_1994 - gdp_1993) / gdp_1993) * 100

print("Percentage GDP growth of India in 1994 compared to 1993:", percentage_growth, "%")
```

 **Find GDP Growth of India from 1960 - 2016**

```python
# Filter the dataset to include only rows for India from 1960 to 2016
india_data = df[(df['Country Name'] == 'India') & (df['Year'] >= 1960) & (df['Year'] <= 2016)]

# Calculate GDP growth for each year
gdp_growth = india_data['Value'].pct_change() * 100

# Remove NaN (resulted from pct_change) and the first row (as it has no previous value to compare with)
gdp_growth = gdp_growth.dropna()
print("GDP Growth of India from 1961 to 2016:")
print(gdp_growth)
```

**Create a new columns named as GDP which shows the GDP growth of each country for that year**

```python
# Group the data by country
grouped = df.groupby('Country Name')

# Calculate the percentage change in GDP for each country
df['Value'] = grouped['Value'].pct_change() * 100

# Drop NaN values (resulted from pct_change) as there's no previous value to compare with for the first year
df = df.dropna(subset=['Value'])

print("Dataset with GDP growth column:")
print(df)
```

**Create a graph of GDP Growth of India from 1960 - 2016**


```python
# Filter the dataset to include only GDP data for India from 1960 to 2016
india_data = df[(df['Country Name'] == 'India') & (df['Year'] >= 1960) & (df['Year'] <= 2016)]

# Calculate GDP growth for each year
india_data['GDP Growth'] = india_data['Value'].pct_change() * 100

# Plot the GDP growth over the years
plt.figure(figsize=(10, 6))
plt.plot(india_data['Year'], india_data['GDP Growth'], marker='o', linestyle='-')
plt.title('GDP Growth of India (1960-2016)')
plt.xlabel('Year')
plt.ylabel('GDP Growth (%)')
plt.grid(True)
plt.xticks(range(1960, 2017, 5))
plt.tight_layout()
plt.show()
```

**GDP of Each Country**

```python
os.mkdir('GDP Individual')

for country_name in df['Country Name'].unique():

    df_pr = df[df['Country Name'] == country_name]
    fig = px.line(df_pr, x = 'Year', y = 'Value', title = country_name + ' GDP Analysis')
    pyo.plot(fig, filename = 'GDP Individual/' + country_name + '.html', auto_open = False)
```

```python
# Create a directory to save individual HTML files if it doesn't exist
os.makedirs('GDP_Individual', exist_ok=True)

# Iterate over unique country names in the DataFrame
for country_name in df['Country Name'].unique():
    # Filter the DataFrame for the current country
    df_country = df[df['Country Name'] == country_name]

    # Create a line plot for the GDP analysis of the current country
    fig = px.line(df_country, x='Year', y='Value', title=country_name + ' GDP Analysis')

    # Save the plot as an HTML file in the 'GDP_Individual' directory
    file_path = os.path.join('GDP_Individual', f'{country_name}_GDP.html')
    pio.write_html(fig, file_path, auto_open=False)

print("Individual HTML files for GDP analysis saved successfully.")
```

**GDP Comparison between Specific Countries**

```python
lst = ['IND','ITA', 'USA', 'CHN']

dfs = []
for i in lst:
    dfs.append(df[df['Country Code'] == i])

    df_pr = pd.concat(dfs, axis = 0)
```

```python
fig = px.line(df_pr, x = 'Year', y = 'Value', title = 'GDP Comparison - ' + '|'.join(lst), color = 'Country Name')

pyo.plot(fig, filename = '|'.join(lst) + '.html')
```

```python
def comare_gdp(lst, isOpen):

    dfs = []
    for i in lst:
        dfs.append(df[df['Country Code'] == i])

        df_pr = pd.concat(dfs, axis = 0)

    fig = px.line(df_pr, x = 'Year', y = 'Value', title = 'GDP Comparison - ' + '|'.join(lst),
                  color = 'Country Name')
    pyo.plot(fig, filename = '|'.join(lst) + '.html', auto_open = isOpen)
```

```python
comare_gdp(['IND','USA'], False)
```
